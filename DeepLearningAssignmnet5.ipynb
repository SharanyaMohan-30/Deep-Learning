```
#Text Generation

# Creating a mapping from unique characters to indices
char2idx = {u:i for i, u in enumerate(vocab)}
#print(char2idx)
idx2char = np.array(vocab)

idx2char = idx2char.tolist()
num_generate = 500
start_string = '<S>'
h_0 = tf.Variable(tf.zeros([1,hidden_size]))
prev_h = h_0
vocab_siz = 68
#print(oldstep)

# input_eval = [char2idx[s] for s in start_string] # For any other symbol


input_eval = [char2idx[start_string]] #For start symbol
input_eval = tf.expand_dims(input_eval, 0)

input_eval = tf.one_hot(input_eval, vocab_size)

text_generated = []

for i in range(num_generate):
  inp = tf.matmul(input_eval,W_xh)
  prev_h = tf.matmul(prev_h, W_hh)
  hidden_layer1 = tf.nn.tanh(inp + prev_h + b_h)
  prev_h = hidden_layer1

  predictions = tf.nn.tanh(tf.matmul(hidden_layer1,W_ho) +b_o)
  #print(predictions)
  predictions = tf.squeeze(predictions, 0)
  #print(predictions.shape)

  # predictions = predictions / temperature
  # print(predictions)
  predicted_id = tf.random.categorical(predictions, num_samples=1)[-1,0].numpy()
  #print(predicted_id)
  input_eval = tf.expand_dims([predicted_id], 0)
  #print(input_eval)
  input_ev = tf.expand_dims([predicted_id], 0)
  #print(input_ev)
  input_eval = tf.one_hot(input_ev, vocab_size)
  #print(input_eval)

  for key, value in idx2char.items(): 
    if input_ev == value: 
      predicted_id = key

  text_generated.append(predicted_id)
start_string + ''.join(text_generated)

```
